### 1. What is name of car that has the highest horsepower?  

```python
highest_hp_car = cars_df['horsepower'].idxmax()
print(highest_hp_car)
```

### 2. How many cars have mpg  35?

```python
#your code here
count_mpg_35 = (cars_df['mpg'] >= 35).sum()
print(count_mpg_35)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
common_origin = cars_df[(cars_df['horsepower'] > 100) & (cars_df['weight'] < 3000)]['origin'].mode()[0]
print(common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
mean_acceleration_japan = round(cars_df[cars_df['origin'] == 'japan']['acceleration'].mean(), 2)
print(mean_acceleration_japan)
```

### 5. Which year had the highest average mpg?

```python
#your code here
highest_avg_mpg_year = cars_df.groupby('model_year')['mpg'].mean().idxmax()
print(highest_avg_mpg_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = cars_df['mpg'].median()
above_median_mpg = cars_df[cars_df['mpg'] > median_mpg]
best_ratio_car = above_median_mpg['horsepower'].div(above_median_mpg['weight']).idxmax()
print("Best hp-to-weight car (above median mpg):", best_ratio_car)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(12,6))
sns.lineplot(
    data=cars_df.reset_index(),
    x='model_year', y='mpg', hue='origin',
    estimator='mean', ci=None
)
plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=cars_df.reset_index(),
    x='horsepower', y='weight',
    hue='origin', size='mpg',
    sizes=(40, 200),
    hue_order=['japan', 'europe', 'usa'],
    palette='Set2',
    alpha=0.7
)
plt.title('Horsepower vs Weight by Origin and MPG')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
